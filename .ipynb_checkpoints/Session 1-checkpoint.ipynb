```python
df_clean = df.copy()
```

```python
df_clean = df_clean[df_clean["RestingBP"] != 0]
```

```python
heartdisease_mask = df_clean["HeartDisease"]==0

cholesterol_without_heartdisease = df_clean.loc[heartdisease_mask, "Cholesterol"]
cholesterol_with_heartdisease = df_clean.loc[~heartdisease_mask, "Cholesterol"]
```

```python
df_clean.loc[heartdisease_mask, "Cholesterol"] = df_clean.loc[heartdisease_mask, "Cholesterol"].replace(0,cholesterol_without_heartdisease.median())
df_clean.loc[~heartdisease_mask, "Cholesterol"] = df_clean.loc[~heartdisease_mask, "Cholesterol"].replace(0,cholesterol_with_heartdisease.median())
```

```python
df_clean[["Cholesterol", "RestingBP"]].describe()
```

```python
df_clean.head()
```

```python
df_clean = pd.get_dummies(df_clean, drop_first=True)
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
```

```python
X = df_clean.drop(["HeartDisease"], axis=1)
y = df_clean["HeartDisease"]
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.15, random_state = 417)

features = [
    "Oldpeak",
    "Sex_M",
    "ExerciseAngina_Y",
    "ST_Slope_Flat",
    "ST_Slope_Up"
]
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
for feature in features:
    knn = KNeighborsClassifier(n_neighbors = 3)
    knn.fit(X_train[[feature]], y_train)
    accuracy = knn.score(X_val[[feature]], y_val)
    print(f"The k-NN classifier trained on {feature} and with k = 3 has an accuracy of {accuracy*100:.2f}%")
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()

X_train_scaled = scaler.fit_transform(X_train[features])
```

```python
X_val_scaled = scaler.transform(X_val[features])
```

```python
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train_scaled, y_train)
accuracy = knn.score(X_val_scaled, y_val)
print(f"Accuracy: {accuracy*100:.2f}")
```

```python
X = df_clean.drop(["HeartDisease"], axis=1)
y = df_clean["HeartDisease"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state = 417)

features = [
    "Oldpeak",
    "Sex_M",
    "ExerciseAngina_Y",
    "ST_Slope_Flat",
    "ST_Slope_Up"
]

scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train[features])
```

```python
grid_params = {"n_neighbors": range(1, 20),
                "metric": ["minkowski", "manhattan"]
              }

knn = KNeighborsClassifier()

knn_grid =GridSearchCV(knn, grid_params, scoring="accuracy")
```

```python
knn_grid.fit(X_train_scaled, y_train)
```

```python
knn_grid.best_score_
```

```python
knn_grid.best_params_
```

```python
knn_grid.best_estimator_
```

```python
from sklearn.metrics import accuracy_score
```

```python
X_test_scaled = scaler.transform(X_test[features])
predictions = knn_grid.best_estimator_.predict(X_test_scaled)
accuracy = accuracy_score(y_test, predictions)
print(f" Model Accuracy on test set: {accuracy*100:.2f}")
```
