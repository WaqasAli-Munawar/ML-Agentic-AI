```python
plt.figure(figsize=(12,15))

for idx,c in enumerate(categorical_cal):
    ax = plt.subplot(4,2,idx+1)
    sns.countplot(x = df[c], ax =ax )

plt.savefig("ccee.png")
plt.show()
```

```python
plt.figure(figsize=(12,15))

for idx,c in enumerate(categorical_cal):
    ax = plt.subplot(4,2,idx+1)
    sns.countplot(x = df[c], ax =ax)
    for container in ax.containers:
        ax.bar_label(container, label_type="center")

# plt.savefig("ccee.png")
plt.show()
```

```python
plt.figure(figsize=(12,15))

for idx,c in enumerate(categorical_cal[:-1]):
    ax = plt.subplot(3,2,idx+1)
    sns.countplot(x = df[c], ax =ax, hue = df['HeartDisease'])
    for container in ax.containers:
        ax.bar_label(container, label_type="center")

# plt.savefig("ccee.png")
plt.show()
```
