```python
input_data = {
  "name": "Waqas",
  "email": "waqasgmail.com",
  "query": 123
}
```

```python
# user_input = UserInput(name = "Waqas", email="waqas@gmail.c",
                       #query = "I lost my password")

UserInput(**input_data)
```

```python
def validate_user_input(input_data):
    try:
        user_input = UserInput(**input_data)
        print("Valid User Input Created")
        print(user_input.model_dump_json(indent=2))
        return user_input

    except ValidationError as e: # e = ValidationError
        # print(type(e))
        for error in e.errors():
            #print(error['loc'][0],":" ,error["msg"])
            print(f"{error['loc'][0]}:{error["msg"]}")
        return None

```

```python
validate_user_input(input_data)
```

```python
from pydantic import Field
from typing import Optional

from datetime import date # year-month-day : 2025-10-05
```

```python
# !pip install pydantic
```

```python
class UserInput(BaseModel):
    name:str 
    email:EmailStr
    query:str
    order_id:Optional[int] = Field(None, 
                                description="5 Digit order number (can not start with 0)",
                                  ge = 10000,
                                  le= 99999)
    purchase_date:Optional[date] = None
```

```python
input_data = {
  "name": "Waqas",
  "email": "waqas@gmail.com",
  "query": "I lost my password",
    
}
validate_user_input(input_data)
```

```python
print(UserInput(**input_data).model_dump_json(indent=2))
```

```python
input_data = {
  "name": "Waqas",
  "email": "waqas@gmail.com",
  "query": "I lost my password",
  "order_id": "02345",
  "purchase_date": date(2015,12,31),
  "Course": "ML",
   "Location": "CCEE"
}

validate_user_input(input_data)
```
